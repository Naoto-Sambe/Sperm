```python
import numpy as np
import pandas as pd
import scanpy as sc
```

```python
sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_header()
sc.settings.set_figure_params(dpi=80, facecolor='white')
```

```python
results_file = 'write/221222_mouse_testis_analysis.h5ad'
```

```python
adata = sc.read_h5ad('write/221222_mouse_testis_processing.h5ad')  
```

```python
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
```

```python
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
sc.pl.highly_variable_genes(adata)
```

```python
adata.raw = adata
```

```python
adata = adata[:, adata.var.highly_variable]
```

```python
adata
```

```python
sc.pp.regress_out(adata, ['total_counts', 'pct_counts_mt'])
```

```python
sc.pp.scale(adata, max_value=10)
```

```python
sc.tl.pca(adata, svd_solver='arpack')
```

```python
sc.pl.pca_variance_ratio(adata, log=True, n_pcs=40)
```

```python
adata.write(results_file)
adata
```

```python
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
```

```python
sc.tl.umap(adata)
```

```python
sc.tl.leiden(adata, resolution = 0.1, key_added = "leiden")
```

```python
sc.set_figure_params(scanpy=True,fontsize=25)
sc.pl.umap(adata, color='leiden')
```

```python
sc.tl.rank_genes_groups(adata, 'leiden', method='wilcoxon', key_added = "wilcoxon")
sc.pl.rank_genes_groups(adata, n_genes=25, sharey=False, key="wilcoxon")
```
